```python
import pandas as pd

data = {'text':['99.73 %','95.45 %','68.27 %'] ,
        'yloc':[.075,.175,.275]
        }

df = pd.DataFrame (data, columns = ['text','yloc'])

print (df)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

# Creating Figure
fig=plt.figure(figsize=(10,5),dpi=100)

mu=10
variance=1
sigma=math.sqrt(variance)

objects = ('mu-3*sigma','mu-2*sigma','mu-sigma','mu','mu+sigma','mu+2*sigma','mu+3*sigma')# u can get these names by doing value_count.index
pos = [mu-3*sigma,mu-2*sigma,mu-sigma,mu,mu+sigma,mu+2*sigma,mu+3*sigma] # this gives something similar to the range fn in python,arange in np is same as range() in python root


# plotting the Normal curve line
x=np.linspace(mu-3*sigma,mu+3*sigma,1000)
plt.plot(x, stats.norm.pdf(x,mu,sigma), color='black',label= 'Mean: {} , Variance: {}'.format(mu,sigma))

# creating Labels for the x-qaxis with xticks
plt.xticks(pos, objects,fontsize=10)


# 68%
lines(p3,p5)

# Filling color-pink in 1 sigma areas
ptx = np.linspace(p3, p5, 10)
pty = stats.norm.pdf(ptx,mu,sigma)
plt.fill_between(ptx, pty, color='pink', alpha=1.0)

#95%
lines(p2,p6)

# Filling color blue in 2 sigma areas
ptx = np.linspace(p5, p6, 10)
pty = stats.norm.pdf(ptx,mu,sigma)
plt.fill_between(ptx, pty, color='blue', alpha=0.3)

ptx = np.linspace(p3, p2, 10)
pty = stats.norm.pdf(ptx,mu,sigma)
plt.fill_between(ptx, pty, color='blue', alpha=0.3)

    
#99.7
lines(p1,p7)

# Filling color-lightgreen in 3 sigma areas
ptx = np.linspace(p6, p7, 10)
pty = stats.norm.pdf(ptx,mu,sigma)
plt.fill_between(ptx, pty, color='lightgreen', alpha=0.3)

ptx = np.linspace(p1, p2, 10)
pty = stats.norm.pdf(ptx,mu,sigma)
plt.fill_between(ptx, pty, color='lightgreen', alpha=0.3)

# Horizontal lines
plt.axhline(y=.05, xmin=.05, xmax=.95,color='black',marker= '|')
plt.axhline(y=.15, xmin=.15, xmax=.85, color='black',marker= '|')
plt.axhline(y=.25, xmin=.25, xmax=.75,color='black',marker= '|')

# Adding text - '99.73 %'-'95.45 %'-'68.27 %' inside the plot
for i in range(len(df)): 
    plt.text(mu-0.35,df['yloc'][i],df['text'][i], color='black', fontsize=20)
       
#Adding Titles and labels    
plt.legend()
plt.ylabel('Probability Density')
plt.title('68-95-99.7 Rule',fontsize=20)
plt.show()
```
